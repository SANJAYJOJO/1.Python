```python
import pandas as pd #importing dataset using pandas,pandas is a library is used to work on datasets with in numarical form
                    #renaming pandas as pd
```

```python
dataset=pd.read_csv("Salary_Data.csv")#we assing the original dataset(salary_data.csv) into a variable  called "dataset"
```

```python
dataset #we calling the variable to show the content in the variable
```

```python
independent=dataset[["YearsExperience"]] #the dataset has two rows,we need to split the input(independent) and output(dependent)
                                         #we assin the input into a variable called "independent", to assin a purticular row form original dataset for a variable we use"[[]]" with the name of the row
```

```python
independent #we calling the variable to show the content in the variable
```

```python
dependent=dataset[["Salary"]] #the dataset has two rows,we need to split the input(independent) and output(dependent)
                              #we assin the output(dependent)in a variable called "dependent"
```

```python
dependent #we calling the variable to show the content in the variable
```

```python
from sklearn.model_selection import train_test_split #Now We Need to split the train and Test Set,for That we using sklearn library in sklearn library callling perticular foloder called "model selection" in that foloder we are importing a particular function called "train_test_split"
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) #the train_test_split has four varialbles, 2 is for input(independent) train and test,another two is for output(dependent) tain and test
```

```python
from sklearn.linear_model import LinearRegression #Here is the Process of Model Creation,From sklearn library we calling a particular foulder called "linear_model" in that linear model we are importing a class called liner regresssoion
regressor=LinearRegression() #we are assin the class(LinearRegression) in variable called regressor
regressor.fit(x_train,y_train) #with help of access operator we are calling particular function called fit method,fit method talks about the substitute process
```

```python
weight=regressor.coef_ #the "corf" is used to find the slop(wight) and assined it for a variable called weight
```

```python
weight #we calling the variable to show the content in the variable
```

```python
bais=regressor.intercept_ #the "intercept" is used to find the bais and assined it for a variable called bais
bais #we calling the variable to show the content in the variable
```

```python
y_pred=regressor.predict(x_test) #we are predecting the value using x_test(x_test is input testing valus)
y_pred #the y_pred is the value that out model is created 
```

```python
from sklearn.metrics import r2_score #To check the Model are accurate we are using validatng parameters one of the valdating parameter is "r_square" from sklearn metrics we importing r2_score(r_square)
r_score=r2_score(y_test,y_pred) #the r2_score law is acutual value minus predected value
```

```python
r_score #if the r_score value is close to 1 we build a perfect model
```
